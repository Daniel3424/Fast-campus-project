#### dict {}
- 순서가 없고 {키:값} 으로 구성되어 있는 데이터 타입

```python
# 선언 : 키는 정수, 문자열 데이터 타입만 사용이 가능
# 인덱스 대신 키를 사용
dic = {
    1: "one",
    "two" : 2,
    "three": [1, 2, 3], 
}
type(dic), dic
```

```python
dic[1], dic["three"]
```

```python
dic["two"] = 123
dic
```

```python
# 솔루션을 리스트로 작성
# 질문 입력 받음
# 솔루션의 갯수에 맞게 랜덤한 index 정수 값을 생성
# index 해당하는 솔루션 리스트의 데이터를 출력
```

```python
# 솔루션을 리스트로 작성
solutions = [
    "무엇을 하던 잘 안될것 이다.",
    "생각지도 않게 좋은 일이 생길것이다.",
    "무엇을 상상하던 그 이상이다.",
]

# 질문 입력 받음
input("질문을 입력하세요. : ")

# 솔루션의 갯수에 맞게 랜덤한 index 정수 값을 생성
idx = random.randint(0, len(solutions) - 1)

# index 해당하는 솔루션 리스트의 데이터를 출력
solutions[idx]
```
